## 12. Describe how list values and string values are identical.

```python
Both strings and lists have lengths: a string's length is the number of characters in the string; a list's length is the number of items in the list.
Each character in a string as well as each item in a list has a position, also called an index.
```

## 13. What's the difference between tuples and lists?

```python
Lists are mutable; they can have values added, removed, or changed. 
Tuples are immutable; they cannot be changed at all. Also, tuples are written 
using parentheses, ( and ), while lists use the square brackets, [ and ].
```

## 17. How do you distinguish between copy.copy() and copy.deepcopy()?

```python
The copy.copy() function will do a shallow copy of a list, while the copy.deepcopy() function will do a deep copy of a list.
That is, only copy.deepcopy() will duplicate any lists inside the list.
```
